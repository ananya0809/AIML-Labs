### About Multiple Linear Regression

Simple Linear Regression: It’s the simplest form of Linear Regression that is used when there is a single input variable for the output variable.

Multiple Linear Regression: It’s a form of linear regression that is used when there are two or more predictors.

Notice that this equation is just an extension of Simple Linear Regression, and each predictor has a corresponding slope coefficient (β).

The first β term (βo) is the intercept constant and is the value of Y in absence of all predictors (i.e when all X terms are 0). It may or may or may not hold any significance in a given regression problem. It’s generally there to give a relevant nudge to the line/plane of regression.

Multiple Linear Regression solves the problem by taking account of all the variables in a single expression.

```python
# Import Pandas Library, used for data manipulation
# Import matplotlib, used to plot our data
# Import nump for mathemtical operations
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Import our WeatherData and store it in the variable weather_data
weather_data = pd.read_csv("WeatherData.csv") 
# Display the data in the notebook
weather_data
```

```python
# Set the features of our model, these are our potential inputs
weather_features = ['Temperature(C)', 'Humidity']

# Set the variable X to be all our input columns: Temperature, Wind Speed and Pressure
X = weather_data[weather_features]

# set y to be our output column: Humidity
y = weather_data.Humidity

# plt.subplot enables us to plot mutliple graphs
# we produce scatter plots for Humidity against each of our input variables

plt.subplot(2,2,1)
plt.scatter(X['Temperature(C)'],y)
plt.subplot(2,2,2)
plt.scatter(X['Humidity'],y)
# plt.subplot(2,2,3)
# plt.scatter(X['Pressure (millibars)'],y)
```

```python
mlr_model = LinearRegression()
mlr_model.fit(X, y)
```

```python
# Import library to produce a 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x1 = X["Temperature(C)"]
# x2 = X["Wind Speed (km/h)"]
x2 = X["Humidity"]

ax.scatter(x1, x2, y, c='r', marker='o')

# Set axis labels
ax.set_xlabel('Temperature(C)')
# ax.set_ylabel('Wind Speed (km/h)')
ax.set_zlabel('Humidity')
```

```python
theta0 = mlr_model.intercept_
theta1,theta2= mlr_model.coef_
theta0,theta1,theta2
```

```python
y_pred = mlr_model.predict([[15, 21],[14, 20]])
y_pred
```
